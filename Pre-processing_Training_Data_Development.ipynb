```python
# import packages

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# load dataset

df = pd.read_csv('esrb_dataset_clean.csv')
df.head()
```

```python
# check missing values

df.isna().sum()
```

```python
X = df.drop(['title','console','esrb_rating'], axis=1) # title and console are not relevant for predictions
y = df['esrb_rating']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```
